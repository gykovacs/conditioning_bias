```python
import numpy as np
import pandas as pd

import sklearn
from sklearn.tree import DecisionTreeClassifier, plot_tree

import matplotlib.pyplot as plt
```

```python
sklearn.__version__
```

```python
data = pd.DataFrame.from_dict([
    {'self-employed': 0, 'income (k)': 80, 'dependents': 1, 'defaults (target)': 1},
    {'self-employed': 0, 'income (k)': 60, 'dependents': 3, 'defaults (target)': 1},
    {'self-employed': 0, 'income (k)': 80, 'dependents': 1, 'defaults (target)': 0},
    {'self-employed': 1, 'income (k)': 50, 'dependents': 2, 'defaults (target)': 0},
    {'self-employed': 1, 'income (k)': 50, 'dependents': 4, 'defaults (target)': 1},
    {'self-employed': 1, 'income (k)': 70, 'dependents': 4, 'defaults (target)': 0}
])

```

```python
print(data.to_latex(index=False))
```

```python
tree = DecisionTreeClassifier(criterion='gini', random_state=5).fit(data[data.columns[:-1]].values, data[data.columns[-1]].values)
plot_tree(tree, feature_names=data.columns, impurity=True, proportion=False, rounded=True, class_names=['pays', 'defaults'], label='all', fontsize=8)
```
