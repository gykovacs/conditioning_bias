```python
import pandas as pd

pd.set_option('display.width', 200)

from sklearn.ensemble import RandomForestRegressor

from flipping_random_forest import FlippingRandomForestRegressor

from datasets import regr_datasets
from core import evaluate
```

```python
data_loaders = regr_datasets['data_loader_function'].values.tolist()
```

```python
scenarios=[{'name': 'original',
            'estimator': RandomForestRegressor,
            'estimator_params': {'n_jobs': 2},
            'multiplier': 1},
            
            {'name': 'flipping',
            'estimator': FlippingRandomForestRegressor,
            'estimator_params': {'n_jobs': 2},
            'multiplier': 1}]
            
compare=[('original', 'flipping')]

validator_params = {'n_repeats': 200,
                    'n_splits': 5,
                    'random_state': 5}
```

```python
results_pdf = evaluate(scenarios=scenarios, 
                        compare=compare, 
                        data_loaders=data_loaders,
                        validator_params=validator_params,
                        score='r2',
                        random_state=5)
```

```python
results_pdf.to_csv('experiment-regression-rf.csv')
```
