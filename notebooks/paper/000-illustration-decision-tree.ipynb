```python
import numpy as np
import pandas as pd

import sklearn
from sklearn.tree import DecisionTreeClassifier, plot_tree

import matplotlib.pyplot as plt
```

```python
sklearn.__version__
```

```python
data = pd.DataFrame.from_dict([
    {'criminal record': 0, 'income (k)': 50, 'years in job': 1, 'loan': 1},
    {'criminal record': 0, 'income (k)': 70, 'years in job': 3, 'loan': 1},
    {'criminal record': 0, 'income (k)': 50, 'years in job': 1, 'loan': 0},
    {'criminal record': 1, 'income (k)': 80, 'years in job': 2, 'loan': 0},
    {'criminal record': 1, 'income (k)': 80, 'years in job': 4, 'loan': 1},
    {'criminal record': 1, 'income (k)': 60, 'years in job': 4, 'loan': 0}
])

```

```python
print(data.to_latex(index=False))
```

```python
tree = DecisionTreeClassifier(criterion='gini', random_state=5).fit(data[data.columns[:-1]].values, data[data.columns[-1]].values)
plot_tree(tree, feature_names=data.columns, impurity=True, proportion=False, rounded=True, class_names=['no loan', 'loan'], label='all', fontsize=8)
```
